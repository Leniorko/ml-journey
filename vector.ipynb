```python
# Setup imports
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
# Configs for pyplot
plt.rcParams["figure.figsize"] = [10, 10]
# plt.rcParams["figure.autolayout"] = True
# Default colors for plots.
# First two are for operands and third is for result.
defaultColors = ['red', 'green', 'blue']

axisMax = 10
axisMin = -10
axisTotal = abs(axisMax) + abs(axisMin)


# NOTE: Apparently I can't figure out proper draw math, so there is 0.5 shortage on Y axis.
# I'll fix it later, probably.
def drawVectorPlot(data: np.ndarray, origin: np.ndarray = np.array([])) -> None:
    # Plot setup
    # If no origin provided fallback to (0, 0) for every vector
    if(origin.size == 0):
        origin = np.zeros((2, data.shape[0]), dtype=int)

    fig, ax = plt.subplots()
    ax.set_xlim(axisMin, axisMax)
    ax.set_ylim(axisMin, axisMax)
    ax.get_yaxis().set_major_locator(MaxNLocator(integer=True, nbins=axisTotal))
    ax.get_xaxis().set_major_locator(MaxNLocator(integer=True, nbins=axisTotal))
    ax.grid(b=True)
    # xy annotation at 0
    plt.quiver(*[[0, axisMin], [axisMin, 0]], [0, axisMax * 2], [axisMax * 2, 0], color="black", angles='xy', scale_units='xy', scale=1)

    print(data[:, 0])

    # Draw test data
    plt.quiver(*origin, data[:, 0], data[:, 1],
               color=defaultColors, angles='xy', scale_units='xy', scale=1)
```

## Vector definition

Vector - something that have magnitude and direction

You can represent vector as: 
$$
\vec{V} = (x, y) = \begin{bmatrix} 
x \\ 
y 
\end{bmatrix}
$$
For example:  $\vec{V} = (5, 0) = \begin{bmatrix} 5 \\ 0 \end{bmatrix}$
That's vector that points to x = 5 and y = 0

```python
data = np.array([[5, 0]])
drawVectorPlot(data)
```

```python
data = np.array([[3, 0], [-1, 2], [4, -1]])
drawVectorPlot(data)
```
