```python
# Setup imports
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
# Configs for pyplot
plt.rcParams["figure.figsize"] = [10, 10]
plt.rcParams["figure.autolayout"] = True

# Default colors for plots.
# First two are for operands and third is for result.
defaultColors = ['red', 'green', 'blue']

axisMax = 10
axisMin = -10
axisTotal = abs(axisMax) + abs(axisMin)

def drawVectorPlot(data: np.ndarray, origin: np.ndarray = np.array([])) -> None:
    # Plot setup
    # If no origin provided fallback to (0, 0) for every vector
    if(origin.size == 0):
        origin = np.zeros((2, data.shape[0]), dtype=int)

    fig, ax = plt.subplots()
    ax.set_xlim(axisMin, axisMax)
    ax.set_ylim(axisMin, axisMax)
    ax.get_yaxis().set_major_locator(MaxNLocator(integer=True, nbins=axisTotal))
    ax.get_xaxis().set_major_locator(MaxNLocator(integer=True, nbins=axisTotal))
    ax.grid(b=True)
    # xy annotation at 0. Use quiver because need to draw below actual data
    plt.quiver(*[[0, axisMin], [axisMin, 0]], [0, axisMax * 2], [axisMax * 2, 0], color="black", angles='xy', scale_units='xy', scale=1)

    # Draw data
    plt.quiver(*origin, data[:, 0], data[:, 1],
               color=defaultColors, angles='xy', scale_units='xy', scale=1)
```

## Vector definition

Vector - something that have magnitude and direction

You can represent vector as: 
$$
\vec{V} = (x, y) = \begin{bmatrix} 
x \\ 
y 
\end{bmatrix}
$$
For example:  $\vec{V} = (5, 0) = \begin{bmatrix} 5 \\ 0 \end{bmatrix}$
That's vector that points to x = 5 and y = 0

```python
data = np.array([[5, 0]])
drawVectorPlot(data)
```

## Sum of vectors

Mathematically sum of vectors is very easy. Simply sum $x$ and $y$ of each vector.
So, here is experimental vectors: 
$\vec{a} = \begin{bmatrix} 6 \\ -2 \end{bmatrix}$ and $\vec{b} = \begin{bmatrix} -4 \\ 4 \end{bmatrix}$

Sum of them will look like this: $\vec{a} + \vec{b} =\begin{bmatrix} 2 \\ 2 \end{bmatrix}$
You can represent graphically by drawing one vector from the end of another. Result will be a vector from the base of first one to the tip of the second one.
Below you can see example.

```python
data = np.array([[6, -2], [-4, 4], [2, 2]])
origins = np.array([[0, 6, 0], [0, -2, 0]])
drawVectorPlot(data, origins)
```

## Subtraction of vectors

Subtraction works same mathematically, but differently graphically.
So if we take vectors from previous step and substract them we will have something like this: 

$\vec{a} - \vec{b} =\begin{bmatrix} 6-(-4) \\ -2-4 \end{bmatrix} = \begin{bmatrix} 10 \\ -6 \end{bmatrix}$

Graphically we will draw vectors with their origin at same place and our result will lie from tip of $\vec{b}$ to tip of $\vec{a}$

```python
data = np.array([[6, -2], [-4, 4], np.subtract([6, -2], [-4, 4])])
# Result moved for more clear view
origins = np.array([[0, 0, -4], [0, 0, 4]])
drawVectorPlot(data, origins)
```

## Vector multiplication with scalar

Simply multiply every part of vector with scalar.
For example: $3\vec{a} = \begin{bmatrix} 3 * 3 \\ -2 * 3 \end{bmatrix} = \begin{bmatrix} 9 \\ -6 \end{bmatrix}$


```python
data = np.array([[3, -2], np.multiply([3, -2], 3)])
# Result moved for more clear view
origins = np.array([[0, -3], [0, 6]])
drawVectorPlot(data, origins)
```
